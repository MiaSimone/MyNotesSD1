# Mean shift and Hierarchical Clustering

Implimentation af AGNES (AGglomerative NESting)

```python
import numpy as np
import pandas as pd

# for drawing dendogram
import scipy.cluster.hierarchy as ch
from scipy.spatial.distance import cdist

# for creating a model
from sklearn.cluster import AgglomerativeClustering
from sklearn import metrics

import matplotlib.pyplot as plt
```

### Data preparation

```python
data = pd.read_csv('../data/shopping-data.csv')
```

```python
data.describe()
```

```python
data.shape
```

```python
data.head(10)
```

```python
X = data.iloc[:, 2:5].values
X
```

## Hierarchical

```python
# Use dendogram to find the optimal number of clusters
plt.figure(figsize=(20,10))

# Linkage is a measure of dissimilarity of sets - distance between sets of observations 
# as a function of the pairwise distances between observations
dendogram = ch.dendrogram(ch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Observations')
plt.ylabel('Euclidean distance')
plt.show()
```

### Train and implement model

```python
# We say n_clusters = 3 because the biggest destance between the horisontal lines is where there are 3 vertical lines, therefor
# 3 clusters
n_clusters = 3
model = AgglomerativeClustering(n_clusters, affinity = 'euclidean', linkage = 'ward')
model.fit(X)
```

```python
Y = model.fit_predict(X)
Y
```

```python
# Visualising the clusters
# Plot the points with color code of the predicted clusters
# viridis - a color map, https://matplotlib.org/users/colormaps.html
# s - marker size, here, the dots size
# c - marker color, here taken from the predicted clusters
plt.scatter(X[:, 0], X[:, 1], s=30, c=Y, cmap='viridis')
plt.title('Discovered Clusters')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

If we remove age from the dataset we will see that 3 or 5 clusters would be better.

## Mean Shift

```python
from sklearn.cluster import  MeanShift, estimate_bandwidth
import seaborn as sns
import plotly.express as pxp
import plotly.graph_objs as gph
```

### Data preparation

```python
data = pd.read_csv('../data/shopping-data.csv', delimiter=',')
print(data)
```

```python
X = data.iloc[:, 3:5].values
X
```

### Train and implement model

```python
# Bandwidth is found automatically with
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=200)
```

```python
# Train mean-shift model with the data frame
msmodel = MeanShift(bandwidth=bandwidth, bin_seeding=True)
#msmodel = MeanShift(bandwidth=bandwidth)
msmodel.fit(X)
```

```python
labels = msmodel.labels_
labels_unique = np.unique(labels)
labels_unique
```

```python
n_clusters_ = len(labels_unique)
n_clusters_
```

```python
# See the clusters and cluster centres
cluster_centers = msmodel.cluster_centers_
cluster_centers
```

```python
# Predict the cluster for all the samples
Y = msmodel.predict(X)
Y
```

```python
# Predict the cluster for all the samples
P = msmodel.predict(X)
P
```

```python
# Visualising the clusters
# Plot the points with color code of the predicted clusters
# viridis - a color map, https://matplotlib.org/users/colormaps.html
# s - marker size, here, the dots size
# c - marker color, here taken from the predicted clusters
fig = plt.figure(figsize=(15,12))
plt.title('Discovered Clusters')
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1],  marker='o', cmap='viridis', c=labels)
ax.scatter(cluster_centers[:,0], cluster_centers[:,1], marker='x', 
           color='red', s=100, linewidth=3, zorder=10)
plt.show()
```

```python
# Generate scatter plot for training data
# colors = list(map(lambda x: '#3b4cc0' if x == 1 else '#b40426' if x == 2 else '#67c614', P))
plt.scatter(X[:,0], X[:,1], c=labels, marker="o", picker=True)
plt.title(f'Estimated number of clusters = {n_clusters_}')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
